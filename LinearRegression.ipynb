```python
# 正规方程线性回归
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from  sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
# 加载数据
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
raw_df.to_csv('boston_house_datasets.csv')
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
print("数据：\n",data)
print("目标值：\n",target)
```

```python
# 数据划分
x_train,x_test,y_train,y_test = train_test_split(data,target,test_size=0.2,random_state=42)
print("训练集的形状:\n",x_train.shape)
print("测试集的个数:\n",x_test.shape)
```

```python
# 特征工程- 标准化
transfer  = StandardScaler()
x_train_transfered = transfer.fit_transform(x_train)
x_test_transfered = transfer.fit_transform(x_test)
print("标准化后的结果：\n",x_train_transfered)
```

```python
# 机器学习
estimator = LinearRegression()
estimator.fit(x_train,y_train)

y_pre = estimator.predict(x_test)
print("预测值为：\n",y_pre)
score = estimator.score(x_test,y_test)
print("准确率是:\n",score)

## 均方误差
ret = mean_squared_error(y_test,y_pre)
print("均方误差是:\n",ret)
```
